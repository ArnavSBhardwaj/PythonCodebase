```python
X = stats.norm(loc=1.0, scale=2.0)
X.stats()
```

```python
x = X.rvs(size=10000) 
np.mean(x), np.var(x) 
```

```python
stats.norm.fit(x) 
```

```python
pdf, t = np.histogram(x, bins=100, normed=True)  #❶
t = (t[:-1] + t[1:]) * 0.5  #❷
cdf = np.cumsum(pdf) * (t[1] - t[0])  #❸
p_error = pdf - X.pdf(t)
c_error = cdf - X.cdf(t)
print ("max pdf error: {}, max cdf error: {}".format(
    np.abs(p_error).max(),
    np.abs(c_error).max()))
```
